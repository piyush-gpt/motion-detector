```python
import cv2
import imutils
import winsound
import threading
```

```python
cv2.namedWindow("window", cv2.WINDOW_NORMAL) 
cv2.resizeWindow("window", 640, 480) 
cap=cv2.VideoCapture(0)
res,start_frame=cap.read()
start_frame=imutils.resize(start_frame,width=500)
start_frame=cv2.cvtColor(start_frame,cv2.COLOR_BGR2GRAY)
start_frame=cv2.GaussianBlur(start_frame,(5,5),0)

alarm=False# to check if already alrm is ringing or not
alarm_mode=False # to switch betwen activation and not activation state of the system of alarm monitoring
alarm_counter=0 # indicates how much movement is allowed before beeping alarm

def beep_alarm():
    global alarm
    for i in range(5): # beep 5 times
        if not alarm_mode:
            break
        print("alarm")
        winsound.Beep(2500,1000)
    alarm=False    


while True:
    res,frame=cap.read()
    frame=imutils.resize(frame,width=500)
    if alarm_mode:
        frame_bw=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        frame_bw=cv2.GaussianBlur(frame_bw,(5,5),0)
        difference=cv2.absdiff(frame_bw, start_frame)
        threshold=cv2.threshold(difference,25, 255,cv2.THRESH_BINARY)[1]
        start_frame=frame_bw
        if threshold.sum()>300:
            alarm_counter+=1
            print(alarm_counter)
        else:
            if alarm_counter>0:
                alarm_counter-=1
        cv2.imshow("window",threshold)
    else:
        cv2.imshow("window",frame)
    if alarm_counter>20:
        if not alarm:
            alarm=True
            threading.Thread(target=beep_alarm).start()# multithreading so that the execution of while loop does not pauses
    key=cv2.waitKey(1)
    if key==ord("t"):
        alarm_mode= not alarm_mode
        alarm_counter=0
    if key==ord("q"):
        alarm_mode= False
        break      
cap.release()
cv2.destroyAllWindows()
```
